#### Data Cleaning

```python
#create a copy of data for processing and manipulation
tdb_copy=telecom_data_base.copy()
```

```python
'''
Use the errors parameter of the to_numeric() function to specify how to handle errors. 
The default value is 'raise', which means that an exception will be raised if any values cannot be converted. 
You can set this parameter to 'coerce', which will replace non-numeric values with NaN values, 
or 'ignore', which will leave non-numeric values as-is.
'''

tdb_copy.TotalCharges=pd.to_numeric(tdb_copy.TotalCharges, errors='coerce')
tdb_copy.isnull().sum()
```

```python
tdb_copy.loc[tdb_copy['TotalCharges'].isnull()==True]
```

```python
tdb_copy.dropna(inplace=True, how='any')

'''
how means: Determine if row or column is removed from DataFrame, when we have
at least one NA or all NA.

* 'any' : If any NA values are present, drop that row or column.
* 'all' : If all values are NA, drop that row or column.
'''
```

```python
tdb_copy.info(verbose=True)
```

```python
print(tdb_copy['tenure'].max())
```

```python
labels=["{0} - {1}".format(i,i+11) for i in range(1, 72, 12)]
tdb_copy['tenure_group']=pd.cut(tdb_copy.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
tdb_copy['tenure_group'].value_counts()
```

```python
'''
Dropping Features that are not relevant for model training.
Customer ID: randomly generated number that will not affect the Churning of a customer
tenure: we have already made a new variable, tenure group.
'''

tdb_copy.drop(columns=['customerID','tenure'], axis=1, inplace=True)
```

##### Univariate Analysis

```python
for i, predictor in enumerate(tdb_copy.drop(columns=['Churn', 'MonthlyCharges', 'TotalCharges'])):
    plt.figure(i)
    sns.countplot(data=tdb_copy, x=predictor, hue='Churn')
```

##### Categorical variables to Dummy variables

##### Dummy trap

##### What is Multicollinearity and How does dummy trap cause Multicollinearity?

```python
tdb_copy_dummies=pd.get_dummies(tdb_copy)
tdb_copy_dummies.head()
```

```python
sns.lmplot(data=tdb_copy_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
#Monthly Charges increases as Total Charges increase: Very Highly Correlated.
```

##### KDE (Kernel Density Estimate) Plot

```python
Mth_charge = sns.kdeplot(tdb_copy_dummies.MonthlyCharges[(tdb_copy_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth_charge = sns.kdeplot(tdb_copy_dummies.MonthlyCharges[(tdb_copy_dummies["Churn"] == 1) ],
                ax =Mth_charge, color="Blue", shade= True)
Mth_charge.legend(["No Churn","Churn"],loc='upper right')
Mth_charge.set_ylabel('Density')
Mth_charge.set_xlabel('Monthly Charges')
Mth_charge.set_title('Monthly charges by churn')

#Churn is HIGH when the Monthly Charges are HIGH.
```

```python
Tot_charge = sns.kdeplot(tdb_copy_dummies.TotalCharges[(tdb_copy_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot_charge = sns.kdeplot(tdb_copy_dummies.TotalCharges[(tdb_copy_dummies["Churn"] == 1) ],
                ax =Tot_charge, color="Blue", shade= True)
Tot_charge.legend(["No Churn","Churn"],loc='upper right')
Tot_charge.set_ylabel('Density')
Tot_charge.set_xlabel('Total Charges')
Tot_charge.set_title('Total charges by churn')

#Churn is HIGH when the Total Charges are LOW.
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=0)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```
